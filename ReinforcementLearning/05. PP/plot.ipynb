```python
import pandas as pd
import json
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
```

```python
def plot(log, x, y, std, title=None):
    _, ax = plt.subplots(figsize=(12, 8))
    
    x = log[x]
    y = log[y]
    std = log[std]

    ax.set_xlabel("â„– of transitions")
    ax.set_ylabel("Mean reward")
    ax.set_title(title)

    plt.plot(x, y)

    plt.fill_between(x, y - std, y + std, alpha=0.5)
    plt.show()
```

```python
with open("experiments/95bd7777-344d-4931-9db5-234671362db8/params.json", "r") as p:
    params = json.load(p)
print(json.dumps(params, indent=4))

log = pd.read_csv("experiments/95bd7777-344d-4931-9db5-234671362db8/log.csv")

plot(log, "step", "predator_mean", "predator_std", title="Predator")
plot(log, "step", "prey_mean", "prey_std", title="Prey")
```
