```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
```

```python
def numd(f: callable, x, eps):
    return (f(x + eps) - f(x - eps)) / (2 * eps)
          
def f(x):
    return np.power(x, 3)

def df(x):
    return 3 * np.power(x, 2)

def plot(x_, prec=np.float64):
    eps_lst = np.logspace(-20, 0, 50, dtype=prec)
    x = np.zeros_like(eps_lst, dtype=prec)
    y = np.zeros_like(eps_lst, dtype=prec)
    for i, eps in enumerate(eps_lst):
        x[i] = eps
        y[i] = np.abs(numd(f, x_, eps) - df(x_))
    plt.loglog(x, y, label=f"x = {x_:.1f}")
    
```

```python
def plot_show(prec=np.float64):
    plt.figure()
    x_lst = np.array([0.1, 0.5, 1], dtype=prec)
    for x in x_lst:
        plot(x, prec)

    plt.axvline(x=np.sqrt(np.finfo(dtype=prec).eps), color="blue", linestyle="--", label="eps ** (1/2)")
    plt.axvline(x=np.finfo(dtype=prec).eps, color="red", linestyle="--", label="eps")
    plt.legend(loc="upper right", ncol=1, bbox_to_anchor=(1.35, 1))
    plt.show()
    
```

```python
plot_show(np.float64)
```

```python
plot_show(np.float32)
```
