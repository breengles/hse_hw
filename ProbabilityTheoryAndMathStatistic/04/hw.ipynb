$\theta^* = \left((k+1)\overline{x^k}\right)^{\frac{1}{k}}$ --- оценка параметра для равномерного распределения

$\theta^* = \left(\frac{\overline{x^k}}{k!}\right)^{\frac{1}{k}}$ --- оценка параметра для экспоненциального распределения

В дальнейшем нам надо понять, какой $k$ лучше всего брать. Так как $U[0,\theta] = \theta \cdot U[0,1]$ и $\exp(\theta)=\theta\cdot\exp(1)$, то будем сэмплировать для стандратных и сранивать с 1.

```python
from Sample import Sample
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
n = 1000
s = 100
k_max = 250

u = Sample(n, s, kind="uniform")
e = Sample(n, s, kind="exp")
```

```python
sns.set()
f = plt.figure()
ax = plt.subplot()
ax.set_xlabel("k")
ax.set_ylabel("RMSE")

ax.plot(range(1, k_max + 1), u.to_plot(k_max), label="Uniform")
ax.legend()
plt.show()
```

```python
f = plt.figure()
ax = plt.subplot()
ax.set_xlabel("k")
ax.set_ylabel("RMSE")

ax.plot(range(1, k_max + 1), e.to_plot(k_max), label="Exponential")
ax.legend()
plt.show()
```

Вывод: для равномерного распределения надо брать k побольше, а для экспоненциального --- поменьше.
